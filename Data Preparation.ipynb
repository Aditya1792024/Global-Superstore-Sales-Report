```python
import pandas as pd
```

```python
# Accessing the orders data.
data = pd.read_excel('global_superstore_2016.xlsx')
```

```python
data.head()
```

```python
# Data Profilling 
```

```python
# 1) Statistical Description of each numerical column in dataset.
data.describe()
```

```python
# 2) Identifying Null values in each column.
data.info()

# Looks like only 'Postal Code' column contains null values and only 19.48% values are non-null. 
# It's better to remove this column.
```

```python
# Removing 'Postal Code' column.
data.drop(columns=['Postal Code'], inplace=True)
```

```python
data.columns
```

```python
# 3) Identifying unique values in each column.
data.nunique()
```

```python
# Changing data type of 'Order Date' column.
data['Order Date'] = data['Order Date'].dt.date
```

```python
# Changing data type of 'Ship Date' column.
data['Ship Date'] = data['Ship Date'].dt.date
```
